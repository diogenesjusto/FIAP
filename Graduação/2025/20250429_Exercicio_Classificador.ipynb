```python
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/refs/heads/master/Gradua%C3%A7%C3%A3o/dados/dados_arvore_titanic.csv")
```

```python
df
```

```python
# separação treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(df[['Age','SibSp']], df['Survived'], test_size=0.3, random_state=42)
```

```python
# modelo de árvore de decisão: classificador
mod = tree.DecisionTreeClassifier()
mod.fit(x_treino, y_treino)
```

```python
# previsão
y_prev = mod.predict(x_teste)
```

```python
# Matriz de Confusão
pd.crosstab(y_prev, y_teste, rownames=['Previsto'], colnames=['Real'], margins=True)
```

```python
# Acurácia (taxa de acertos) => ACC = (TP+TN)/TOT
accuracy_score(y_teste, y_prev)
```

```python
# Precision = TP/(TP+FP)
precision_score(y_teste, y_prev)
```

```python
# Recall = TP/(TP+FN)
recall_score(y_teste, y_prev)
```
