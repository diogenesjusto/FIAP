```python
# 1. Carga de bibliotecas
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# 2. Carga de dados
df = pd.read_csv('https://raw.githubusercontent.com/diogenesjusto/FIAP/master/SHIFT/Data/pib.csv')
df
```

```python
# 3. Separação de treino e teste
treino = df.iloc[0:132]
teste = df.iloc[132:138]


# 4. Regressão linear
Y_treino = treino.iloc[:,[2]]

#X = ['BRP']  # Regressão Linear Simples
#X = ['PIBi1', 'PIBi2'] # Modelo Autoregressivo de ordem 2
X = ['PIBi1', 'PIBi2', 'PIBi12'] # Modelo Autoregressivo de ordem 2 + ajuste de caude
X_treino = treino.loc[:,X]

modelo = LinearRegression()
modelo.fit(X_treino, Y_treino)

print(modelo.score(X_treino, Y_treino))
X_teste = teste.loc[:,X]

Y_prev = modelo.predict(X_teste)
Y_real = teste.loc[:,['PIB']]
Erro = Y_prev - Y_real
abs(Erro).mean() # Abs = valor absoluto ou módulo, ou seja, sem sinal. Ou ainda, raiz quadrada do quadrado | Mean = média. Juntando tudo: RMSE
```

# Tratamento de valores nulos em Séries Temporais

```python
import numpy as np
```

```python
# forçar entrada de nulo
df.at[2,'PIB']=np.nan
df.at[3,'PIB']=np.nan
df
```

```python
# Preenchimento de valores nulos de uma S.T. com forward fill - isto é, utilizar último valor existente para preencher próximos nulos
df['PIBtff'] = df['PIB'].fillna(method='ffill')
```

```python
df
```

```python
# Preenchimento de valores nulos a partir de métodos de interpolação
df['PIBtinterpolate'] = df['PIB'].interpolate()
```
