```python
# Carga de dados
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/dados/train.csv")
```

```python
# Matriz Contingência
pd.crosstab(df['Survived'], 1, margins=True)
```

```python
# Probabilidade de sobrevivência no Titanic? (amostra = train.csv) => P = X/T
P_Sobrev = 342/891
P_Sobrev
```

```python
# Probabilidade morte = 1 - P_Sobrev
P_Morte = 1 - P_Sobrev
P_Morte
```

```python
# Matriz Contingência
pd.crosstab(df['Survived'], df['Sex'], margins=True)
```
