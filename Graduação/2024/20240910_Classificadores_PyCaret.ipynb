```python
# Importar biblioteca pycaret
from pycaret.classification import *
```

```python
# Carga de dados
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/dados/train.csv")
```

```python
# Configuração pycaret
s = setup(df, target='Survived')
```

```python
# Construção de modelos e comparação
best = compare_models()
```

```python
# Detalhes do algoritmo com melhores métricas
print(best)
```

```python
# Análise das métricas e avaliação do melhor modelo
evaluate_model(best)
```
