```python
# Exercício 09/04 - parte 1

# Carga de bibliotecas
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error as MAE
import statsmodels.api as sm

# Carga de dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/SHIFT/Data/pib.csv")

# Separação de treino e teste
treino = df.iloc[0:126]
teste = df.iloc[126:138]

# Regressão
Y_treino = treino.iloc[:,[2]]
#X = ['BRL', 'BRP'] # Regressão Linear Simples
# Modelo AutoRegressivo
X = ['PIBi1', 'PIBi2', 'PIBi4', 'PIBi12']
X_treino = treino.loc[:,X]
mod = LinearRegression()
mod.fit(X_treino, Y_treino)

print(mod.score(X_treino,Y_treino))

# Cálculo de erro com python
X_teste = teste.loc[:,X]
Y_prev = mod.predict(X_teste)
Y_real = teste.loc[:,['PIB']]
Erro = Y_prev - Y_real

# Erro de previsão
print(MAE(Y_real, Y_prev))

mod_sm = sm.OLS(Y_treino, X_treino).fit()
print(mod_sm.summary())
```

```python
# Exercício 1 - 02/04/2024

df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/desafio_baba.csv")

# Como construir colunas com lag
df['venda1'] = df['venda'].shift(1)
df['venda2'] = df['venda'].shift(2)
df['venda3'] = df['venda'].shift(3)
df['venda4'] = df['venda'].shift(4)
df['venda5'] = df['venda'].shift(5)
df['venda6'] = df['venda'].shift(6)
df['venda7'] = df['venda'].shift(7)

#df.boxplot(['venda'], by='weekday')

# Separação de treino e teste
treino = df.iloc[7:334]
teste = df.iloc[334:365]

Y_treino = treino.loc[:,['venda']]
# Modelo AutoRegressivo
X = ['venda1', 'venda3', 'venda4']
X_treino = treino.loc[:,X]
mod = LinearRegression()
mod.fit(X_treino, Y_treino)

# Análise de coeficientes
print(mod.coef_)
print(mod.intercept_)

mod_sm = sm.OLS(Y_treino, X_treino).fit()
print(mod_sm.summary())

# Cálculo de erro com python
X_teste = teste.loc[:,X]
Y_prev = mod.predict(X_teste)
Y_real = teste.loc[:,['venda']]
Erro = Y_prev - Y_real

# Erro de previsão
print(MAE(Y_real, Y_prev))
```

```python
# Exercício 2 - 09/04/2024

# Carga de dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/SHIFT/Data/pib.csv")

# Contruis Lags de BRL e BRP
df['BRPi1'] = df['BRP'].shift(1)
df['BRPi2'] = df['BRP'].shift(2)
df['BRLi1'] = df['BRL'].shift(1)
df['BRLi2'] = df['BRL'].shift(2)

# Separação de treino e teste
treino = df.iloc[2:126]
teste = df.iloc[126:138]

# Regressão BRL
Y_treino = treino.loc[:,['BRL']]
# Modelo AutoRegressivo
X = ['BRLi1', 'BRLi2']
X_treino = treino.loc[:,X]
mod = LinearRegression()
mod.fit(X_treino, Y_treino)
# Previsão em teste
X_teste = teste.loc[:,X]
BRL_prev = mod.predict(X_teste)

# Regressão BRP
Y_treino = treino.loc[:,['BRP']]
# Modelo AutoRegressivo
X = ['BRPi1', 'BRPi2']
X_treino = treino.loc[:,X]
mod = LinearRegression()
mod.fit(X_treino, Y_treino)
# Previsão em teste
X_teste = teste.loc[:,X]
BRP_prev = mod.predict(X_teste)

dfPrev = pd.DataFrame(BRL_prev, columns=['BRL'])
dfPrev['BRP'] = BRP_prev
```

```python
# Modelo Multivariado BRL e BRP
Y_treino = treino.loc[:,['PIB']]
# Modelo Multivariado
X = ['BRL', 'BRP']
X_treino = treino.loc[:,X]
mod = LinearRegression()
mod.fit(X_treino, Y_treino)

# Previsão em teste
#X_teste = teste.loc[:,X]
X_teste = dfPrev
Y_prev = mod.predict(X_teste)

# Cálculo de erro com python
Y_real = teste.loc[:,['PIB']]
Erro = Y_prev - Y_real

# Erro de previsão
print(MAE(Y_real, Y_prev))
```
