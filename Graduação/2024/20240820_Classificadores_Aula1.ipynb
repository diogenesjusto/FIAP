```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score

df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/Fraud_Study_Enrich.csv")
```

```python
df.shape
```

```python
df.head(3)
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(df[['COD_NOTA_SERVICO', 'DT_MES_EXECUCAO']], df['RESULTADO_N'], test_size=0.32, random_state=25)

x_treino.shape
```

```python
x_teste.shape
```

```python
# Construção do modelo "classificador"
mod = tree.DecisionTreeClassifier()
mod = mod.fit(x_treino, y_treino)
```

```python
# Previsão em teste
y_prev = mod.predict(x_teste)
y_prev.shape
```

```python
y_teste.shape
```

```python
y_prev
```

```python
y_teste
```

```python
# Cálculos de erro de previsão de um classificador
# 1.o) acurácia
accuracy_score(y_teste, y_prev)
```

```python
# 2.o) Matriz de confusão
pd.crosstab(y_prev, y_teste, margins=True)
```
