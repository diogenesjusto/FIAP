```python
# Importação das bibliotecas | regressor ou classificador
from pycaret.classification import *
import pandas as pd
```

```python
# Carga de dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/dados/fraud_study_enrich.csv")
```

```python
# Remoção da coluna RESULTADO)N
df = df.drop(['RESULTADO_N'], axis=1)
```

```python
# Setup do conjunto de dados
# ANÁLISE DESCRITIVA DOS DADOS
s = setup(df, target='RESULTADO')
```

```python
# Variáveis Dummy / One Hot Encoding
## criação de dummies para uma variável categórica binária
pd.get_dummies(df[['OBITO']])
```

```python
## criação de dummies para uma variável categórica com mais de 2 categorias
pd.get_dummies(df[['SEXO']])
```

```python
# Treinamento de modelos/aplicação de diferentes algoritmos
best = compare_models()
```

```python
# Análise do melhor modelo
evaluate_model(best)
```
