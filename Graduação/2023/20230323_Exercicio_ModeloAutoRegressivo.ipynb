```python
# 1. Carga de bibliotecas e base de dados
import pandas as pd
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/pib.csv")
```

```python
# 2. Análise Exploratória
df
```

```python
# 3. Separação de Treino e Teste
# Geração de amostra em uma série temporal (falaremos mais sobre isto adiante)
# através de uma janela temporal
treino = df.iloc[0:124]
teste  = df.iloc[125:137]
```

```python
# 4. Construção das variáveis para regressão linear

# TREINO
# Variável X - localizamos a variável BRP na 4a posição indicial 
# Exercício: experimentar com diferentes variáveis e diferentes combinações de variáveis, avaliando o resultado do modelo preditivo

X_treino = treino.loc[:,['PIBi1', 'PIBi2']]

# Variável Y - PIB
Y_treino = treino.iloc[:,[2]]
```

```python
# 5. Construção regressão
mod = LinearRegression()
mod.fit(X_treino,Y_treino)
```

```python
# 6. Previsão em teste
X_teste = teste.loc[:,['PIBi1', 'PIBi2']]
Y_teste = teste.iloc[:,[2]]

Y_prev = mod.predict(X_teste)
```

```python
# 7. Análise do Erro de Previsão
# Uso da métrica MSE
mse = mean_squared_error(Y_teste, Y_prev)

mse
```
