```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, ConfusionMatrixDisplay

df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/Fraud_Study_Enrich.csv")
```

```python
# Criação de Dummy para tratamento de uma variável categórica
#dummies_Sex = pd.get_dummies(df['Sex'])
#f['SexFemale'] = dummies_Sex['female']

# Separação de Amostras de treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(df[['COD_NOTA_SERVICO','FATURADO']], df['RESULTADO_N'], test_size=0.32, random_state=25)
```

```python
x_teste.shape
```

```python
# Modelo de árvore de decisão
#mod = tree.DecisionTreeClassifier()
# Modelo Random Forest
#mod = RandomForestClassifier()
# Modelo SVM
mod = svm.SVC()
mod = mod.fit(x_treino, y_treino)
```

```python
# Previsão em teste
y_prev = mod.predict(x_teste)

accuracy_score(y_prev, y_teste)
```

```python
precision_score(y_prev, y_teste)
```

```python
recall_score(y_prev, y_teste)
```

```python
# Matriz de Confusão (Confusion Matrix)
pd.crosstab(y_prev, y_teste, margins=True)
```
