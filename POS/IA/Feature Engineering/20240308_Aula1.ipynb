```python
# Carga de bibliotecas
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
# Carga de dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/dados/mtcars.csv")
df
```

```python
# Gráfico de dispersão
df.plot.scatter(x='wt', y='mpg')
```

```python
# Ajuste de escala "manual"
#df['wt_t'] = df['wt']*6

# Ajuste de escala utilizando função do sklearn
scaler = MinMaxScaler(feature_range=(0,1))
df['mpg01'], df['wt01'] = scaler.fit_transform(df[['mpg']]), scaler.fit_transform(df[['wt']])
```

```python
# Execução do K-Means
k = KMeans(n_clusters=3, random_state=1).fit(df[['wt01','mpg01']])
df['cluster'] = k.labels_
df.plot.scatter(x='wt', y='mpg', c='cluster', colormap='Accent')
```
