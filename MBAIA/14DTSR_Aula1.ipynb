```python
# Carga de bibliotecas
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
# Carga de dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/dados/mtcars.csv")
df
```

```python
# Gráfico de dispersão
df.plot.scatter(x="wt", y="mpg")
```

```python
# Ajuste de escala - normalização
# 1.o fazendo o cálculo manual - nova variável wt_t
df['wt_t'] = df['wt']*7

# 2.o utilizando função
scaler = MinMaxScaler(feature_range=(0,1))
df['mpg01'],df['wt01'] = scaler.fit_transform(df[['mpg']]), scaler.fit_transform(df[['wt']])
```

```python

# Execução K-Means
k = KMeans(n_clusters=4, random_state=1).fit(df[['wt01','mpg01']])

k.labels_
```

```python
# Gráfico com sinalização de cores
df.plot.scatter(x="wt", y="mpg", c=k.labels_, colormap='Accent')
```
