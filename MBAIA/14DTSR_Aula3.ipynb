# Dados "temporais" - Time Series

```python
# Dados do Itau Unibanco PIBIU
# + dados de tráfego da ABCR: https://melhoresrodovias.org.br/
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# 2. Carga de dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/SHIFT/Data/pib.csv")
df
```

```python
# 2. Construção das variáveis com Lag
df['PIBi1_'] = df['PIB'].shift(1)
df['PIBi2_'] = df['PIB'].shift(2)
```

```python
# 3. Separação de treino e teste
# Dados temporais(série temporal) => amostra por janela temporal
treino = df.iloc[0:132]
teste = df.iloc[132:138]
```

```python
#4. Regressão Linear
Y_treino = treino.iloc[:,[2]]

#X = ['BRP'] # RLS, exógena BRP

#X = ['PIBi1', 'PIBi2'] # AR2, modelo autoregressivo de ordem 2
#X = ['PIBi1', 'PIBi2','PIBi3'] # AR3, modelo autoregressivo de ordem 3
#X = ['PIBi1', 'PIBi2','PIBi3', 'PIBi4'] # AR4, modelo autoregressivo de ordem 4
#X = ['PIBi1', 'PIBi2','PIBi3', 'PIBi4', 'PIBi12'] # AR4, modelo autoregressivo de ordem 4+ajuste sazonal-> tendência

X = ['BRP', 'D1', 'D2', 'D3', 'D4', 'D5', 'D6', 'D7', 'D8', 'D9', 'D10', 'D11'] # Modelo Multivariado com tratamento de sazonalidade através de dummies
X_treino = treino.loc[:,X]

mod = LinearRegression()
mod.fit(X = X_treino, y = Y_treino)

print(mod.score(X_treino, Y_treino))
```

```python
print(mod.coef_) # coef angular ou a
print(mod.intercept_) # coef linear ou b
```

```python
#5. Previsão em teste
X_teste = teste.loc[:,X]
Y_prev = mod.predict(X_teste)

Y_prev
Y_real = teste.iloc[:,[2]]
Erro = Y_prev - Y_real
abs(Erro).mean()
```
