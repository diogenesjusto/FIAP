```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/refs/heads/master/dados/mtcars.csv")
```

```python
# Transformação de variáveis com normalização (ajuste de escala)
df[['mpg2', 'wt2']] = MinMaxScaler(feature_range=(0,1)).fit_transform(df[['mpg', 'wt']])
```

```python
df
```

```python
# Gráfico de dispersão entre wt (peso) e mpg (inv. consumo)
df.plot.scatter(x='wt2', y='mpg2')
```

```python
# Execução do K-Means
k = KMeans(n_clusters=3, random_state=1 )
k.fit(df[['wt2', 'mpg2']])
```

```python
df['cluster'] = k.labels_

df.plot.scatter(x='wt', y='mpg', c='cluster', colormap='plasma')
```

```python
# 1. Carga Bibliotecas e Dados
import pandas as pd
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

dfTit = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/SHIFT/Data/titanic/train.csv")

# 2. Tratamento de dados
# Dummy de "Sex"
dummy_sex = pd.get_dummies(dfTit['Sex'])
dfTit['SexFemale'] = dummy_sex['female']
# Dummy de "Pclass"
dummy_pclass = pd.get_dummies(dfTit['Pclass'], prefix="Class")
dfTit['Class_1'] = dummy_pclass['Class_1']
dfTit['Class_2'] = dummy_pclass['Class_2']

# Preenchimento de nulos para idades
ageMed = dfTit['Age'].mean()
dfTit['Age'] = dfTit['Age'].fillna(ageMed)

# 3. Separação Treino Testo / Modelo / Previsão
# Separação treino: ajuste fino/calibragem => validação cruzada (cross validation)
x_treino, x_teste, y_treino, y_teste = train_test_split(dfTit[['SexFemale', 'Age', 'Class_1', 'Class_2']], dfTit['Survived'], test_size=0.25, random_state=1)

# Preparação/Modelo
# mod = tree.DecisionTreeClassifier() # árvores de decisão
# Ajuste fina/calibragem => hyper parameter tunning
# mod = RandomForestClassifier(random_state=3, n_estimators=150) # Random Forest
mod = svm.SVC() # SVM como um classificador

# DICA DE PYTHON (muitos perguntam): Se você desejar UNIR dois dataframes utilize o comando=> X = pd.concat([X, X], axis=1) <= no exemplo "dentro" dos sinais > e <, o dataframe X é "duplicado" lado a lado. Faça o teste
mod.fit(x_treino ,y_treino )
# Previsão
y_prev = mod.predict(x_teste)
# Confusion Matrix
pd.crosstab(y_prev, y_teste, margins=True)
```
