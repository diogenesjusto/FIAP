# BDS FIAP Aula 2: Estatística Descritiva

```python
# Carga da biblioteca pandas para trabalhar com dados no python
import pandas as pd
# Carga do arquivo externo csv para um dataframe do pandas
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/diamonds.csv")
```

```python
# Visualizando as primeiras 5 linhas (cabeçalho) do data frame
df.head()
```

```python
# Contando a quantidade de linhas e colunas do dataframe
df.shape
```

```python
# Estatística descritivas
# Média aritmética
df[['price']].mean()
```

```python
# Mediana
df[['price']].median()
```

```python
# Desv. Padrão (STandard Deviation)
df[['price']].std()
```

# Geração de Amostras

```python
# Gerando uma amostra de tamanho 3000 e verificando se ela representa a população
# iloc - fç do pandas que filtra um dataframe por [linhas,colunas]. Utilizar : significas desde o item 1 até o último item
a1 = df.iloc[0:3000,:]
```

```python
a1
```

```python
a1[['price']].mean()
```

```python
a1[['price']].median()
```

```python
a1[['price']].std()
```

```python
# Conclusão: como a estatísticas descritivas são muito diferentes entre população e amostra, podemos concluir que a amostra não representa a população.
```

```python
# Nova tentativa:
a2 = df.iloc[3000:6000,:]
a2
```

```python
print(a2[['price']].mean())
print(a2[['price']].median())
print(a2[['price']].std())
```

```python
# 3a tentativa: captar aleatoriamente dados de todas as "partes" do dataframe. Ou seja, vamos reordenar aleatoriamente o dataframe.
a3 = df.sample(3000, random_state=1)
a3
```

```python
a3[['price']].mean()
```

```python
a3[['price']].median()
```

```python
a3[['price']].std()
```
