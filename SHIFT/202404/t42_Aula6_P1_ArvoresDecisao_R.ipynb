Visualização Gráfica de Árvores utilizando R no Colab
Obs: antes de iniciar, clicar em Conexão -> Alterar tipo de ambiente de execução -> Mudar de Python 3 para R

```r
# Instalação de biblioteca
install.packages("party")
```

```r
# Carga de biblioteca em R
library("party")
```

```r
df <- read.csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/2021/S1/dados_arvore.csv")
```

```r
# ctree é um comando da biblioteca party que cria árvores de decisão
# Sintaxe: ctree(Y~X1+...Xn, data=dataframe_origem)
# Obs: função as.factor transforma uma variável em categórica
mod <- ctree(Survived~Age+as.factor(Pclass)+as.factor(Embarked), data=df )
```

```r
# Apresentação gráfica da árvore
plot(mod, type="simple")
```
