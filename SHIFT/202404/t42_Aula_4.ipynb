```python
# 1. Carga de bibliotecas
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# 2. Carga de dados
pib = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/SHIFT/Data/pib.csv")
```

```python
# 3. Tratamento e transformação de dados
# Lag1
pib['PIBi1_'] = pib['PIB'].shift(1)
# Preenchimento de valores nulos em Séries Temporais
pib['PIBi1_f']= pib['PIBi1_'].fillna(method='backfill')
```

```python
# 4. Separação de treino e teste
treino = pib.iloc[0:126]
teste = pib.iloc[126:138]
```

```python
# 5. Regressão Linear
Y_treino = treino.iloc[:,[2]]

#X = ['BRL'] # RLS
#X = ['BRL','BRP'] # RLM
#X = ['PIBi1','PIBi2'] # AR2
#X = ['PIBi1','PIBi2','PIBi12'] # AR2+Saz12
X = ['BRL', 'BRP','D1','D2','D3','D4','D5','D6','D7','D8','D9','D10'] # RLM+dummies
X_treino = treino.loc[:,X]

mod = LinearRegression()
mod.fit(X_treino, Y_treino)
```

```python
# 6. Previsão em teste
X_teste = teste.loc[:,X]
Y_real = teste.loc[:,['PIB']]

Y_prev = mod.predict(X_teste)

Erro = Y_prev - Y_real
ErroAbsMedio = np.mean(abs(Erro))   # Abs = valor absoluto = raiz quadrada do quadrado
ErroAbsMedio
```
