```python
# Carga de bibliotecas e dados
import pandas as pd
from sklearn.cluster import K
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP-On-Line-Graduation/main/datasets/%20mtcars.csv")
```

```python
# Análise de dispersão
df.plot.scatter(x='wt', y='mpg')
```

```python
# Transformação de Variáveis

# Ajustes de escalas - Normalização de variáveis

# Manual: Mesma escala entre Mpg e WT
df['mpg2'] = df['mpg']/5

# Utilizando função MinMax
scaler = MinMaxScaler(feature_range=(0,1))
df['mpg3'], df['wt3'] = scaler.fit_transform(df[['mpg']]), scaler.fit_transform(df[['wt']])
```

```python
df
```

```python
# Execução do K-Means
k = KMeans(n_clusters=4, random_state=1).fit(df[['mpg3','wt3']])
```

```python
# Identificação dos clusters => k.labels_
df['cluster'] = k.labels_
# Gráfico de dispersão, diferenciando os objetos (veículos) por 3 cores diferentes (parâmetro "c")
df.plot.scatter(x='wt', y='mpg', c='cluster', colormap='plasma')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
dfTit = pd.read_csv('https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/dados_arvore_titanic.csv')
# Transformações de dados

# 1. Criação de dummies
dummy_sex = pd.get_dummies(dfTit['Sex'])
dfTit['SexFemale'] = dummy_sex['female']
# 2. Preenchimento de nulos para idades
ageMed = dfTit['Age'].mean()
dfTit['Age'] = dfTit['Age'].fillna(ageMed)
x_treino, x_teste, y_treino, y_teste = train_test_split(dfTit[['SexFemale', 'Age', 'Pclass','Parch', 'SibSp']], dfTit['Survived'], test_size=0.25, random_state=1)

# Construir modelo de árvore de decisão
#mod = tree.DecisionTreeClassifier()
# Construir modelo RandomForest
#mod = RandomForestClassifier(random_state=1, n_estimators=2000)
# Construir modelo SVM
mod = svm.SVC()

# Treinamento
mod.fit(x_treino, y_treino)
# Previsão em teste
y_pred = mod.predict(x_teste)
# Métricas de erro de classificadores
print(accuracy_score(y_teste, y_pred))
print(precision_score(y_teste, y_pred))
print(recall_score(y_teste, y_pred))
```
