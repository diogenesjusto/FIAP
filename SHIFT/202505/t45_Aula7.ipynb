```python
# Cargas de bibliotecas
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
# Carga de dados
df = pd.read_csv('https://raw.githubusercontent.com/diogenesjusto/FIAP-On-Line-Graduation/main/datasets/%20mtcars.csv')
```

```python
# Análise exploratória
df.plot.scatter(x='wt', y='mpg')
```

```python
# Transformação de variáveis - cálculo para 1 variável
df['wt6'] = df['wt']*6   # Red. de escala em WT

# Ajuste de escala para n-variáveis
scaler = MinMaxScaler(feature_range=(0,1))
df[['wt01', 'mpg01']] = scaler.fit_transform(df[['wt', 'mpg']])
```

```python
# Execução do KMeans
k = KMeans(n_clusters=4, random_state=1)
k.fit(df[['wt01', 'mpg01']])
```

```python
df['cluster'] = k.labels_
df.plot.scatter(x='wt', y='mpg', c='cluster', colormap='plasma')
```
