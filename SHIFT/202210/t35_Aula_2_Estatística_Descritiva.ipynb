```python
#Utilizando o # podemos colocar um comentário em um código que não será interpretado
# Carregar biblioteca pandas
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/diamonds.csv")
```

```python
df.head
```

```python
# Cálculo da média (aritm)
df[['price']].mean()
```

```python
# Cálculo da mediana
df[['price']].median()
```

```python
# Cálculo de desvio padrão
df[['price']].std()
```

```python
# Geração de amostras
# Queremos gerar uma amostra de tamanho 3 mil linhas
a1 = df.iloc[0:2999, :]
# iloc = filtro de linhas e colunas por índice; 0:2999 é filtro da linha 1 até linha 3000; : é um filtro aberto, ou seja, da 1a até última coluna
a1.head
```

```python
# cálculo estatísticas descritivas da amostra
a1[['price']].mean()
```

```python
a1[['price']].median()
```

```python
a1[['price']].std()
```

```python
# Vamos tentar gerar uma nova amostra pois a a1 não representa a população
a2 = df.iloc[3000:5999, :]
a2[['price']].mean()
```

```python
a2[['price']].median()
```

```python
a2[['price']].std()
```

```python
# Geração de amostra aleatória
a3 = df.sample(3000, random_state=2)
a3.head
```

```python
a3[['price']].mean()
```

```python
a3[['price']].median()
```

```python
a3[['price']].std()
```

```python
# Histograma
df[['price']].hist()
```

```python
a1[['price']].hist()
```

```python
a2[['price']].hist()
```

```python
a3[['price']].hist()
```

```python
# Histogramas por segmentos de dados 
df.hist(by='color', column='price')
```

```python
# Boxplot
df.boxplot(['price'])
```

```python
# Boxplot por categoria
df.boxplot(['price'], by='cut')
```

```python
# Cálculo de decis ou percentis
import numpy as np
df[['price']].quantile(np.arange(0,1,0.1))
```

```python
a3[['price']].quantile(np.arange(0,1,0.1))
```

```python
# Gráfico de dispersão - ScatterPlot
df.plot.scatter(x='price', y='carat')
```

```python
# Coeficiente de Correlação Linear
df[['price', 'carat']].corr()
```
