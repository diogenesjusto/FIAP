```python
# 1. Carga de bibliotecas
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# 2. Carga de Dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/SHIFT/Data/pib.csv")
```

```python
# 3. Separação de treino e teste
treino = df.iloc[0:132]   # iloc filtra linhas
teste = df.iloc[132:138]
```

```python
# 4. Regressão Linear
Y_treino = treino.iloc[:,[2]]

#X = ['BRL']  # RLS - reg linear simples
# X = ['BRL', 'BRP']  # RLM - reg linear múltipla
# X = ['PIBi1', 'PIBi2']   # AR2 - Modelo AutoRegressivo ordem 2
# X = ['PIBi1', 'PIBi2','PIBi4', 'PIBi12']   # AR4 - O3 + S12
X = ['BRL', 'BRP', 'D1', 'D2', 'D4', 'D5', 'D6', 'D7', 'D8', 'D9', 'D11']  # RLM - reg linear múltipla com Dummies de sazonalidade

X_treino = treino.loc[:,X]

modelo = LinearRegression()
modelo.fit(X_treino, Y_treino)
```

```python
# 5. Previsão em teste
X_teste = teste.loc[:,X]
Y_prev = modelo.predict(X_teste)
Y_real = teste.loc[:, ['PIB']]

Erro = Y_prev - Y_real
ErroAbsMedio = abs(Erro).mean()

print(ErroAbsMedio)
```

```python
# Como ver os coeficientes da equação
modelo.intercept_ # Coeficiente Linear ou coef. intercepto
```

```python
modelo.coef_ # Coeficientes angulares
```

```python
# R2
print(modelo.score(X_treino, Y_treino) )
```

```python
# p_value
import statsmodels.api as sm
mod_sm = sm.OLS(Y_treino, X_treino).fit()
print(mod_sm.summary())
```
