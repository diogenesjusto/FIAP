# Aprendizagem Não Supervisionada
## K-Means

```python
# Carga de bibliotecas
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
# Carga de dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/refs/heads/master/dados/mtcars.csv")
```

```python
# Gráfico de dispersão entre Wt(peso) e mpg (inv. consumo)
df.plot.scatter(x='wt', y='mpg')
```

```python
# Transf. variáveis (manual)
df['wt6'] = df['wt'] * 6
df.plot.scatter(x='wt6', y='mpg')
```

```python
# Transf. variáveis com normalização (ajuste de escala)
df[['mpg01','wt01']] = MinMaxScaler().fit_transform(df[['mpg', 'wt']])

df.plot.scatter(x='wt01', y='mpg01')
```

```python
# Execução do K-Means
k = KMeans(n_clusters=5, random_state=1)
k.fit(df[['wt01', 'mpg01']])
```

```python
df['cluster'] = k.labels_
df.plot.scatter(x='wt', y='mpg', c='cluster', colormap='plasma')
```

# 2a Parte - Outros algoritmos

```python
# 1. Carga Bibliotecas e Dados
import pandas as pd
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

dfTit = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/SHIFT/Data/titanic/train.csv")

# 2. Tratamento de dados
# Dummy de "Sex"
dummy_sex = pd.get_dummies(dfTit['Sex'])
dfTit['SexFemale'] = dummy_sex['female']
# Dummy de "Pclass"
dummy_pclass = pd.get_dummies(dfTit['Pclass'], prefix="Class")
dfTit['Class_1'] = dummy_pclass['Class_1']
dfTit['Class_2'] = dummy_pclass['Class_2']
# Preenchimento de nulos para idades
ageMed = dfTit['Age'].mean()
dfTit['Age'] = dfTit['Age'].fillna(ageMed)

# 3. Separação Treino Testo / Modelo / Previsão
# Separação treino: ajuste fino/calibragem => validação cruzada (cross validation)
x_treino, x_teste, y_treino, y_teste = train_test_split(dfTit[['SexFemale', 'Age', 'Class_1', 'Class_2']], dfTit['Survived'], test_size=0.25, random_state=1)

# Preparação/Modelo
#mod = tree.DecisionTreeClassifier() # árvores de decisão
#mod = RandomForestClassifier(random_state=1, n_estimators=5000, verbose=1) # Random Forest
mod = svm.SVC() # SVM

# Ajuste de parâmetros de um algoritmos: hyper parameter tunnnig

# Treinamento
mod.fit(x_treino ,y_treino )
# Previsão
y_prev = mod.predict(x_teste)
# Confusion Matrix
pd.crosstab(y_prev, y_teste, margins=True)
print(accuracy_score(y_teste, y_prev))
print(recall_score(y_teste, y_prev))
print(precision_score(y_teste, y_prev))
```
