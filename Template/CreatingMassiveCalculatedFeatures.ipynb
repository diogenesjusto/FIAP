```python
import pandas as pd
di = pd.read_csv('https://raw.githubusercontent.com/diogenesjusto/FIAP/master/Gradua%C3%A7%C3%A3o/dados/diamonds.csv')
di
```

```python
ndi = di.min()

id = 1

for i in ndi:
  nc = 'min_'+ndi.index[id-1]
  di[nc]=i
  id=id+1
```
